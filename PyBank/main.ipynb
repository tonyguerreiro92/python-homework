```python
# -*- coding: UTF-8 -*-
"""PyBank Homework."""

# Import Pandas Library to read the CSV file
import csv

budget_data_csv = [];
with open('budget_data.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        budget_data_csv.append(row)

months = []
profits_lossses = []

for i, element in enumerate(budget_data_csv):
    months.append(element['Date'])
    profits_lossses.append(int(element['Profit/Losses']))
    

# Create all the needed variables to calculate and parse the CSV file
number_of_months = 0
profit_losses_total = 0
profit_losses_average = 0
past_revenue = 0
revenues_array = []
months_array = []
profit_array = []
greatest_increase_month = ""
greatest_increase_amount = 0
greatest_decrease_month = ""
greatest_decrease_amount = 0

# Iterate through all the rows of the file and extract the index as well
for i, month in enumerate(months):
    # Calculate the number of months
    number_of_months = number_of_months + 1
    
    # Calculate the profit losses total
    profit_losses_total = profit_losses_total + profits_lossses[i]
    
    # Calculate revenue change, past revenue, and add to the profit arrays
    revenue_change = float(profits_lossses[i])- past_revenue
    past_revenue = float(profits_lossses[i])
    profit_array.append(float(profits_lossses[i]))
    
    # Determine the greatest increase and greatest decrease amount and month
    
    # If the revenue change is greater than the greatest increase amount, set new greatest increase amount as the current one
    if revenue_change > greatest_increase_amount:
        greatest_increase_amount= revenue_change
        greatest_increase_month = months[i]

    # If the revenue change is lesser than the greatest decrease amount, set new greatest decrease amount as the current one
    if revenue_change < greatest_decrease_amount:
        greatest_decrease_amount= revenue_change
        greatest_decrease_month = months[i]

# Determine the revenues of each month by substracting the profit of the next month minus the profit of the current month
for i in range(len(profit_array) - 1):
    revenues_array.append(profit_array[i + 1] - profit_array[i])

# Calculate the average of the profit/losses by adding all of them together and dividing it by the number of samples
profit_losses_average = sum(revenues_array)/len(revenues_array)
     
# Print in the local python file
print("Financial Analysis")
print("----------------------------")
print(f"Total Months: {number_of_months}")
print(f"Total: ${profit_losses_total}")
print(f"Average  Change: ${round(profit_losses_average, 2)}")
print(f"Greatest Increase in Profits: {greatest_increase_month} (${greatest_increase_amount})")
print(f"Greatest Decrease in Profits: {greatest_decrease_month} (${greatest_decrease_amount})")

# Create text output file with the report
with open('financial_analysis.txt', 'w') as financial_analysis:
    financial_analysis.write("Financial Analysis")
    financial_analysis.write("\n")
    financial_analysis.write("----------------------------")
    financial_analysis.write("\n")
    financial_analysis.write(f"Total Months: {number_of_months}")
    financial_analysis.write("\n")
    financial_analysis.write(f"Total: ${profit_losses_total}")
    financial_analysis.write("\n")
    financial_analysis.write(f"Average  Change: ${round(profit_losses_average, 2)}")
    financial_analysis.write("\n")
    financial_analysis.write(f"Greatest Increase in Profits: {greatest_increase_month} (${greatest_increase_amount})")
    financial_analysis.write("\n")
    financial_analysis.write(f"Greatest Decrease in Profits: {greatest_decrease_month} (${greatest_decrease_amount})")
    financial_analysis.write("\n")
```
